```python
from datasets import load_dataset
import datasets
import re

# Suppress Hugging Face logs
datasets.utils.logging.set_verbosity_error()

# Load MedMCQA dataset
dataset_path = "/home/hao/colm/datasets/MedMCQA"
dataset = load_dataset("medmcqa", cache_dir=dataset_path)

# Choose split
split = "train"

def preprocess_medmcqa(dataset, split):
    """Extract and clean questions and corresponding answers from MedMCQA dataset."""
    if split not in dataset:
        raise ValueError(f"Error: Split '{split}' not found. Available splits: {list(dataset.keys())}")
    
    processed_data = []
    for item in dataset[split]:
        question = item.get("question")
        options = item.get("options", [])  # Multiple-choice options
        correct_index = item.get("cop", -1)  # Correct option index
        
        if question and options and 0 <= correct_index < len(options):
            correct_answer = options[correct_index]
            processed_data.append({"question": question, "answer": correct_answer})
    
    if not processed_data:
        raise ValueError("Error: No valid 'question' and 'answer' fields found in the dataset.")
    
    # Text cleaning
    for item in processed_data:
        item["question"] = re.sub(r"[^a-zA-Z0-9.,!? ]", "", item["question"]).strip()
        item["answer"] = re.sub(r"[^a-zA-Z0-9.,!? ]", "", item["answer"]).strip()
    
    return processed_data

# Process the text data
try:
    processed_samples = preprocess_medmcqa(dataset, split)
    print(f"Processed {len(processed_samples)} question-answer pairs.")
    
    # Print a few samples
    for i, sample in enumerate(processed_samples[:5]):
        print(f"Sample {i+1}:")
        print(f"Question: {sample['question']}")
        print(f"Answer: {sample['answer']}")
        print("-")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from datasets import load_dataset
import datasets
import re

# Suppress Hugging Face logs
datasets.utils.logging.set_verbosity_error()

# Load MedMCQA dataset
dataset_path = "/home/hao/colm/datasets/MedMCQA"
dataset = load_dataset("medmcqa", cache_dir=dataset_path)

# Choose split
split = "train"

# Debug: Print a sample entry to check structure
if split in dataset:
    print("Sample entry structure:", dataset[split][0])

def preprocess_medmcqa(dataset, split):
    """Extract and clean questions and corresponding answers from MedMCQA dataset."""
    if split not in dataset:
        raise ValueError(f"Error: Split '{split}' not found. Available splits: {list(dataset.keys())}")
    
    processed_data = []
    for item in dataset[split]:
        question = item.get("question", "").strip()
        options = [item.get("opa", ""), item.get("opb", ""), item.get("opc", ""), item.get("opd", "")]
        correct_index = item.get("cop", -1)  # Correct option index
        
        # Validate question, options, and correct answer
        if question and all(options) and 0 <= correct_index < len(options):
            correct_answer = options[correct_index]
            processed_data.append({"question": question, "answer": correct_answer})
    
    if not processed_data:
        raise ValueError("Error: No valid 'question' and 'answer' fields found in the dataset.")
    
    # Text cleaning
    for item in processed_data:
        item["question"] = re.sub(r"[^a-zA-Z0-9.,!? ]", "", item["question"]).strip()
        item["answer"] = re.sub(r"[^a-zA-Z0-9.,!? ]", "", item["answer"]).strip()
    
    return processed_data

# Process the text data
try:
    processed_samples = preprocess_medmcqa(dataset, split)
    print(f"Processed {len(processed_samples)} question-answer pairs.")
    
    # Print a few samples
    for i, sample in enumerate(processed_samples[:5]):
        print(f"Sample {i+1}:")
        print(f"Question: {sample['question']}")
        print(f"Answer: {sample['answer']}")
        print("-")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import openai
import pandas as pd
import re
from datasets import load_dataset

# Load the MedMCQA dataset from Hugging Face
dataset = load_dataset("openlifescienceai/medmcqa")

# OpenAI Client
client = openai.OpenAI()

# Define the GPT Prompt
prompt = """
Generate a list of 50 medically relevant (Cause, Effect) pairs based on common clinical knowledge from the MedMCQA dataset.
The pairs should be concise, medically accurate, and formatted strictly as:
(Cause, Effect)

Example:
(Smoking, Lung Cancer)
(High Blood Pressure, Stroke)
(Vitamin D Deficiency, Weak Bones)
(Obesity, Type 2 Diabetes)

Now generate 50 more such medically relevant (Cause, Effect) pairs:
"""

# Call GPT to generate causal pairs
try:
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )
    
    # Extract response content
    gpt_output = response.choices[0].message.content.strip()

    # Extracting cause-effect pairs using regex
    causal_pairs = []
    for match in re.findall(r"\(([^,]+),\s*([^,]+)\)", gpt_output):
        cause, effect = match
        causal_pairs.append((cause.strip(), effect.strip()))

    # If API fails, use a fallback list
    if not causal_pairs:
        print("⚠️ GPT returned an empty response. Using fallback data.")
        causal_pairs = [
            ("Smoking", "Lung Cancer"),
            ("High Blood Pressure", "Stroke"),
            ("Obesity", "Type 2 Diabetes"),
            ("Excessive Alcohol", "Liver Cirrhosis"),
            ("Chronic Stress", "Hypertension"),
            ("Long-term NSAID Use", "Gastric Ulcers"),
            ("Hepatitis B Virus", "Hepatocellular Carcinoma"),
            ("Helicobacter Pylori Infection", "Peptic Ulcer Disease"),
            ("Untreated Diabetes", "Diabetic Neuropathy"),
            ("Sedentary Lifestyle", "Metabolic Syndrome"),
            ("Vitamin B12 Deficiency", "Peripheral Neuropathy"),
            ("Chronic Hyperlipidemia", "Atherosclerosis"),
            ("Heart Valve Infection", "Endocarditis"),
            ("Prolonged Hypertension", "Heart Failure"),
            ("Severe Dehydration", "Acute Kidney Injury"),
            ("Untreated Appendicitis", "Peritonitis"),
            ("Chronic Alcoholism", "Wernicke’s Encephalopathy"),
            ("Excessive Sugar Intake", "Insulin Resistance"),
            ("Long-Term Bed Rest", "Pressure Ulcers"),
            ("Severe Burns", "Hypovolemic Shock"),
        ]

    # Convert to DataFrame
    df = pd.DataFrame(causal_pairs, columns=["Cause", "Effect"])
    
    # Save to CSV
    output_path = "medMCQA_generated_causal_pairs.csv"
    df.to_csv(output_path, index=False)

    print(f"\n✅ Successfully generated {len(df)} causal pairs and saved to {output_path}")

except Exception as e:
    print(f"❌ Error generating causal pairs: {e}")
```
